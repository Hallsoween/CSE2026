```python
x_pos=0
y_pos=0
a_pos=0
b_pos=0
c_pos=0
d_pos=0
u_pos=0
v_pos=0
x_direct=-1
y_direct=1
a_direct=1
b_direct=-1
c_direct=1
d_direct=1
u_direct=-1
v_direct=-1
fig,ax=plt.subplots(figsize=(13,10))
ax.quiver(x_pos, y_pos, x_direct, y_direct)
ax.quiver(a_pos, b_pos, a_direct, b_direct)
ax.quiver(c_pos, d_pos, c_direct, d_direct)
ax.quiver(u_pos, v_pos, u_direct, v_direct)
ax.set_title('Quiver plot with four arrows')
plt.show()
```

```python
fig,ax1=plt.subplots(figsize=(14,8))
x=np.arange(0,2.2,0.2)
y=np.arange(0,2.2,0.2)
X,Y=np.meshgrid(x,y)
x_direct=np.cos(X)*Y
y_direct=np.sin(Y)*Y
n=-2

color=np.sqrt(((y_direct-n)/2)*2+((x_direct-n)/2)*2)
ax1.quiver(X,Y,x_direct,y_direct,color,alpha=0.8)
ax1.set_aspect('equal')
ax1.set_title('Meshgrid Function')
```
